```python
from dummy_data import *
from random import choice
```

```python
# Energy Usage

average_monthly_electricity_bill = choice(avg_electric_bill('small'))
average_monthly_natural_gas_bill = choice(avg_gas_bill('small'))
average_monthly_fuel_bill = choice(avg_transport_cost('small'))

def energy_usage(average_monthly_electricity_bill, average_monthly_natural_gas_bill, average_monthly_fuel_bill):
    '''
    This function calculates the energy usage of a company.

    Inputs:
    - average_monthly_electricity_bill: the average monthly electricity bill in euros
    - average_monthly_natural_gas_bill: the average monthly natural gas bill in euros
    - average_monthly_fuel_bill: the average monthly fuel bill for transportation in euros
    
    Outputs:
    - energy: (float)
    '''
    # energy_usage = (average_monthly_electricity_bill * 12 * 0.0005) + (average_monthly_natural_gas_bill * 12 * 0.0053) + (average_monthly_furl_bill * 12 * 2.32)
    return (average_monthly_electricity_bill * 12 * 0.0005) + (average_monthly_natural_gas_bill * 12 * 0.0053) + (average_monthly_fuel_bill * 12 * 2.32)

eu = energy_usage(average_monthly_electricity_bill, average_monthly_natural_gas_bill, average_monthly_fuel_bill)
print(f"Your Company's energy usage: {eu:.2f} kgCO2")
```

```python
# Waste 

waste_per_month_kg = choice(waste_generated_dummy('small'))
waste_recycled_percent = choice(recycled_waste('small'))

def waste_generated(waste_per_month_kg, waste_recycled_percent):
    '''
    This function calculates the CO2 emissions from waste generation

    Inputs:
    waste_per_month_kg: waste generated per month in kilograms
    waste_recycled_percent: percentage of waste recycled or composed
    
    Outputs:
    waste: CO2
    '''
    # waste = (waste_per_month_kg * 12 * 0.57 - waste_recycled_percent)
    return (waste_per_month_kg * 12 * 0.57 - waste_recycled_percent)

waste = waste_generated(waste_per_month_kg, waste_recycled_percent)
print(f"Your Company's waste generation: {waste:.2f} kgCO2")
```

```python
# Business Travel 

employee_travel = choice(travel_kms('small'))
average_fuel_efficiency = choice(vehicle_fuel_efficiency('small'))

def business_travel(employee_travel, average_fuel_efficiency):
    '''
    This function calculates the CO2 emissions from business travel.

    Inputs:
    employee_travel (int): The number of kilometers traveled by employees for business purposes.
    average_fuel_efficiency (int): The average fuel efficiency of the vehicles used for business travel in liters per 100 kilometers.

    Outputs:
    float: The CO2 emissions from business travel in kilograms of CO2.
    '''
    return (employee_travel * 1 / average_fuel_efficiency * 2.31)

bt = business_travel(employee_travel, average_fuel_efficiency)
print(f"Your Company's business travel: {bt:.2f} kgCO2")
```

```python
import matplotlib.pyplot as plt

# defining labels
activities = ['Energy Usage', 'Waste Generated', 'Business Travel']

# portion covered by each label
slices = [eu, waste, bt]

# color for each label
colors = ['r', 'y', 'g']

# plotting the pie chart
plt.pie(slices, labels = activities, colors=colors, 
        startangle=90, shadow = True, explode = (0, 0, 0),
        radius = 1.2, autopct = '%1.1f%%')

# plotting legend
plt.legend()

# showing the plot
plt.show()
```
